```python
myData = data[data['지역명']=='경기도 용인시 수지구'].copy()
```

```python
myData.info()
```

```python
myData.drop(columns=['지역교육청명', '학교급명', '설립구분명', '주야간명', '제외여부', '제외사유' ], inplace=True)
```

```python
testDf = myData.copy()

testDf
```

```python
testDf['상위고등 진학 총계'] = testDf['(특수목적고)과학고진학합계(명)']+testDf['(특수목적고)외고ㆍ국제고진학합계(명)']+testDf['(자율고)자율형공립고진학합계(명)']
```

```python
testDf['상위고등 진학 총계']
```

```python
dfTotal = testDf.copy()
dfTotal.sort_values('상위고등 진학 총계', axis=0, ascending=False, inplace=True)
```

```python
dfTotal.head(10)
```

```python
totalYear = []

for i in range(2015, 2019) :
    print(i)
    temp = dfTotal[dfTotal['기준년도']==i].copy()
    totalYear.append(temp)
```
